```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# simple custom dataset
# features =[1,2,3,4,5,6,7,8]
# targets = [10,25,30,40,46,60,70,80]
data=pd.read_csv('dataset.csv')
features=data['features']
targets=data['targets']
```

```python
# load dataset using pandas
df=pd.DataFrame({'features':features,'targets':targets})
print(df.head(5))
```

```python
# plotting data points
plt.scatter(df['features'],df['targets'],c='red')
plt.xlabel('features')
plt.ylabel('targets')
plt.title('features vs targets ')
plt.show()
```

```python
# fit the model corresponding to features and targets
model = LinearRegression()
model.fit(df[['features']],df[['targets']])
```

```python
#model predictions
predictions=model.predict(df[['features']])
```

```python
# lets see the best fit line
plt.scatter(df['features'],df['targets'],c='red')
plt.plot(df['features'],predictions,c='green',label='Best Fit Line')
plt.legend()
plt.show()
```

```python
# evaluation use diff metrics to understand behaviour for complex datasets
mse = mean_squared_error(df['targets'],predictions)
print('Mean Squared Error:',mse)

mae = mean_absolute_error(df['targets'], predictions)
print('Mean Absolute Error:', mae)
```
